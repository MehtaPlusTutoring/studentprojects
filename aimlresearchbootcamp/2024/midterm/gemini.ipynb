Import Beautiful Soup Libraries

```python
from urllib.request import Request, urlopen
from bs4 import BeautifulSoup
```

```python
import pandas as pd
```

```python
import chromadb
from chromadb.db.base import UniqueConstraintError
from chromadb.utils import embedding_functions
client = chromadb.PersistentClient(path="db")  # data stored in 'db' folder
em = embedding_functions.SentenceTransformerEmbeddingFunction()
collection = client.create_collection(name='langchain', embedding_function=em)
```

Drop All Columns Except for Name and URL

```python
df = pd.read_csv('/content/drive/MyDrive/gifts.csv')
url_column = df['Link to Museum']
name_column = df['Name']
df
```

Function That Returns Description From URL

```python
def get_description(url):
  req = Request(
      url=url,
      headers={'User-Agent': 'Mozilla/5.0'}
  )
  webpage = urlopen(req).read()
  webpage = BeautifulSoup(webpage, 'html.parser')
  results = webpage.find_all("div", class_="panel-panel-inner")
  description = results[1].find_all("div", class_="field-type-text-with-summary")
  cleantext = BeautifulSoup(description[0].text, "lxml").text.replace("\n", "").replace("\xa0", "")
  return cleantext
```

Create A List Of The URL and IDs

```python
url_list = url_column.tolist()
id_list = name_column.tolist()

description_list = []
for i in url_list:
  description_list.append(get_description(i))

print(description_list)
```

Add the Appropriate Variables to the Collection

```python
collection.add(
        documents = description_list,
        ids = id_list
)
print(collection.get())
```

Initialize the Prompt

```python
question = "Why is the Gouro Sculpture (Symbolic of Femininity) significant?"
```

Print Results from Collection

```python
results = collection.query(
    query_texts=question,
    n_results=1
)
```

Configure Gemini and Generate Response

```python
from google.generativeai.types import HarmCategory, HarmBlockThreshold
```

```python
import os

import google.generativeai as genai

genai.configure(api_key=API_KEY)

generation_config = {
  "temperature": 1,
  "top_p": 0.95,
  "top_k": 64,
  "max_output_tokens": 8192,
  "response_mime_type": "text/plain",
}

model = genai.GenerativeModel(
  model_name="gemini-1.5-flash",
  generation_config=generation_config,
  safety_settings = {
        HarmCategory.HARM_CATEGORY_SEXUALLY_EXPLICIT: HarmBlockThreshold.BLOCK_ONLY_HIGH,
    }

)

chat_session = model.start_chat(
  history=[
  ]
)
response = chat_session.send_message(f'Given {results["documents"][0]}, answer {question}')
print(response.text.replace("*", ""))
```
